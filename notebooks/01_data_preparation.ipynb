```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the synthetic dataset
file_path = '../data/synthetic_star_dataset.csv'
data = pd.read_csv(file_path)
```

```python
data.info()
```

```python
data.head()
```

```python
# Plot histograms for Teff, logg, and Lum
features = ['Teff', 'logg', 'Lum']
plt.figure(figsize=(15, 5))

for i, feature in enumerate(features, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data[feature], kde=True, bins=75, color='blue')
    plt.title(f'{feature} Distribution')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Plot distributions of Teff, logg, Lum for each Star Type
for feature in ['Teff', 'logg', 'Lum']:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x=feature, hue='Star_Type', kde=True, bins=100, palette="tab10")
    plt.title(f'{feature} Distribution by Star Type')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend(title='Star Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

```python
# Count the number of samples per star type
class_counts = data['Star_Type'].value_counts()

# Plot the class distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.title('Star Type Distribution')
plt.xlabel('Star Type')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45)
plt.show()

print("Class Distribution:")
print(class_counts)
```
