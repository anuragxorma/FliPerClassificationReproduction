```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sys
import os

# Get the parent directory
parent_dir = os.path.dirname(os.getcwd())

# Add 'src' and 'data' folders to the system path
sys.path.append(os.path.join(parent_dir, 'src'))
```

```python
from data_preprocessing import generate_light_curve, compute_psd
```

```python
# Load synthetic data
data = pd.read_csv("../data/synthetic_star_dataset.csv")
```

```python
# Simulate light curves for a subset of stars (visualization purposes)
for idx, row in data.sample(5).iterrows():
    time, flux = generate_light_curve(row['Teff'], row['Star_Type'])
    freq, psd = compute_psd(time, flux)

    # Plot PSD
    plt.figure(figsize=(10, 5))
    plt.loglog(freq, psd, label=f"{row['Star_Type']} ({row['Teff']}K)")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Power Spectral Density")
    plt.title(f"PSD for {row['Star_Type']}")
    plt.legend()
    plt.show()

print("Light curve and PSD visualization complete.")
```
